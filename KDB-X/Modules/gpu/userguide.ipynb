# KDB-X GPU Edition User Guide

The `gpu` module provides the ability to accelerate q-based KDB-X workloads in GPU enabled environments.

The module provides a number of APIs allowing users to build KDB-X workloads which can leverage GPU compute for the following operations:
- Joins: `aj` (asof joins)
- Sorts and Searches: `iasc`, `asc`, `xasc`, and `bin`
- Functional `selects` including binary operations, where clause, and group by aggregations

KDB-X workloads can be built from existing GPU enabled KDB-X ecosystems.  Operations can be peformed on data resident across both CPU and GPU devices.  The module provides efficient APIs to manage data transfers across devices.

In addition, the module provides a consise representation of the data, whether resident across both CPU and GPU devices directly from KDB-X processes.
